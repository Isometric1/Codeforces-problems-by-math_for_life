```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
import math
! pip install catboost
from catboost import CatBoostRegressor
! pip install phik
import phik
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/content/house/train.csv')
train,test = train_test_split(df, train_size=0.8)
main_test = pd.read_csv('/content/house/test.csv')
main_test['GarageArea'] = main_test['GarageArea'].fillna(0)
```

```python
X = ['YrSold','LotArea','GarageArea','GrLivArea','TotRmsAbvGrd']
y = ['SalePrice']
```

```python
test.head()
```

```python
model = CatBoostRegressor(loss_function = 'RMSE', eval_metric= 'MAPE', random_state= 36, early_stopping_rounds= 100,learning_rate = 0.1, verbose = 100)
model.fit(train[X],train[y])
test['SalePrice_p'] = model.predict(test[X])
mean_absolute_error(test['SalePrice'],test['SalePrice_p'])
```

```python
grid = GridSearchCV(CatBoostRegressor(),param_grid = {'learning_rate':[0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.5],'eval_metric':['MAPE','MAE','RMSE'],'loss_function':['RMSE']})
grid.fit(train[X],train[y])
test['SalePrice_p1'] = grid.predict(test[X])
mean_absolute_error(test['SalePrice'],test['SalePrice_p1'])
```

```python
grid.best_params_
```

```python
main_test['SalePrice'] = grid.predict(main_test[X])
main_test[['Id','SalePrice']].to_csv('/content/Gradient1.csv')
```
