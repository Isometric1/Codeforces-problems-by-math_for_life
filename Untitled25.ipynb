```python
df_test = pd.read_csv('/content/novozymes/test.csv',sep = ',')
train_updates = pd.read_csv('/content/novozymes/train_updates_20220929.csv',sep = ',')
df_train = pd.read_csv('/content/novozymes/train.csv')
df_train.drop(columns = 'data_source', inplace = True)
df_test.drop(columns = 'data_source', inplace = True)
for seq_id in train_updates.seq_id:
    df_train = df_train.drop(index = seq_id)
df_train.sort_values('tm',ascending = False, inplace = True)
df_train.reset_index(inplace = True)
df_train.drop(columns = 'seq_id', inplace = True)
# For df_train
lst = []
for i in range(len(df_train)):
    lst.append(len(df_train['protein_sequence'][i]))
df_train['length'] = lst



# For df_test
lstf = []
for i in range(len(df_test)):
    lstf.append(len(df_test['protein_sequence'][i]))
df_test['length'] = lstf
letters = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
for _ in letters:
  df_train[_] = df_train['protein_sequence'].str.count(_)
  df_test[_] = df_test['protein_sequence'].str.count(_)
df_train.head(10)
```

## XGBRegressor

```python
df_train[(df_train['length']<240)&(df_train['length']>200)&(df_train['pH'] == 8)]
```

```python
X = df_train.drop(columns = ['tm','protein_sequence','index'])

y = df_train['tm']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 11)
# parameters tuned separately
model1 = xgb.XGBRegressor(n_estimators = 700, max_depth = 5)
model1.fit(X_train, y_train)
predictions1 = model1.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions1)
```

```python
mean_absolute_percentage_error(y_test, predictions1)
```

```python
Z = df_test.drop(columns = ['protein_sequence','seq_id'])

pred = model1.predict(Z)
kaggle = df_test.copy()
kaggle['tm'] = pred
kaggle.head(30)
```
